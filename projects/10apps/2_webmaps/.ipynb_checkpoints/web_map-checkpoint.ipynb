```python
import folium
import pandas as pd
```

```python
data = pd.read_csv('Volcanoes.txt')

lat=list(data['LAT'])
lon=list(data['LON'])
elev = list(data['ELEV'])
volc = list(data['NAME'])

html = """
Volcano name:<br>
<a href="https://www.google.com/search?q=volcano+%s" target="_blank">%s</a><br>
Height: %s m
"""

data.head()
```

```python
map = folium.Map(location=[35.5,-115.1], zoom_start=6,tiles='Stamen Terrain')
```

```python
fg = folium.FeatureGroup(name='My Map')

for lt, ln, v, el in zip(lat, lon, volc, elev):
    iframe = folium.IFrame(html=html %(v,v,el), width=200, height=100)
    fg.add_child(folium.Marker(location=[lt,ln],popup=folium.Popup(iframe),icon=folium.Icon(color='green')))
    
map.add_child(fg)
map
```

```python
map.save('Map1.html')
```
