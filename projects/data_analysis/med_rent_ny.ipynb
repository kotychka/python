```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.express as px
import matplotlib as mpl
import matplotlib.pyplot as plt

df = pd.read_csv('D:/Cloud/Nastya/IT/python/median_price.csv') # reading data
df_ny = df.loc[df['City']=='New York']

## dynamic since 2010 to 2016, by years means

## dynamic since 2010 to 2016
df_ny_new = df_ny.set_index('RegionName').transpose().reset_index().iloc[5:] #  transposing columns into rows
# draw figure    
fig = go.Figure()
for region in df_ny['RegionName']:
	fig.add_trace(go.Scatter(y=df_ny_new[region], x=df_ny_new['index'], mode='lines+markers', name=region))
fig.update_layout(title={'text': "The dynamic of average prices of 1 bedroom apartments in New York"})
fig.show()
# ## dynamic for 2014
# df_ny_2014 = df_ny.set_index('RegionName').transpose().reset_index().iloc[-33:-21] #  transposing columns into rows
# # draw lines    
# fig = go.Figure()
# for region in df_ny['RegionName']:
# 	fig.add_trace(go.Scatter(y=df_ny_2014[region], x=df_ny_2014['index'], mode='lines+markers', name=region))
# fig.update_layout(title={'text': "The dynamic of average prices of 1 bedroom apartments in ny during 2014"})
# fig.show()
## dynamic for 2015
df_ny_2015 = df_ny.set_index('RegionName').transpose().reset_index().iloc[-21:-9] #  transposing columns into rows
# draw lines    
fig = go.Figure()
for region in df_ny['RegionName']:
	fig.add_trace(go.Scatter(y=df_ny_2015[region], x=df_ny_2015['index'], mode='lines+markers', name=region))
fig.update_layout(title={'text': "The dynamic of average prices of 1 bedroom apartments in ny during 2015"})
fig.show()
## dynamic for 2016
df_ny_2016 = df_ny.set_index('RegionName').transpose().reset_index().iloc[-9:] #  transposing columns into rows
# draw lines    
fig = go.Figure()
for region in df_ny['RegionName']:
	fig.add_trace(go.Scatter(y=df_ny_2016[region], x=df_ny_2016['index'], mode='lines+markers', name=region))
fig.update_layout(title={'text': "The dynamic of average prices of 1 bedroom apartments in ny during 2016"})
fig.show()
## dynamic since 2014 to 2016
df_ny_2014_16 = df_ny.set_index('RegionName').transpose().reset_index().iloc[-33:] #  transposing columns into rows
# draw lines
fig = go.Figure()
for region in df_ny['RegionName']:
	fig.add_trace(go.Scatter(y=df_ny_2014_16[region], x=df_ny_2014_16['index'], mode='lines+markers', name=region))
fig.update_layout(title={'text': "The dynamic of average prices of 1 bedroom apartments in New York since 2014 to 2016"})
fig.show()

mean_rent = df_ny_new.mean()
min_rent = df_ny_new.min()
max_rent = df_ny_new.max()
print(mean_rent, min_rent, max_rent)
# df

### Conlusion: the prices in more expensive areas slowly went down since late 2014 till mid 2015. 
### And then slowly went up. More chiper places have sharp ups and downs but eventually go up.
```
