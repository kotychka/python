```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = [0, 1,2,3,4]
y = [0, 2,4,6,8]

## resize the graph
plt.figure(figsize=(5,3), dpi=100)

# plt.plot(x,y, label='2x', color='green', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='red')

## shorthand notation
#fmt = '[color][marker][line]'
plt.plot(x,y, 'r*--', label='2x')

## line #2
x2 = np.arange(0,4.5,0.5)
# plt.plot(x2, x2**2, 'g', label='^2')
plt.plot(x2[:6], x2[:6]**2, 'g', label='^2') # for graphs with prediction or smth like this
plt.plot(x2[5:], x2[5:]**2, 'g--', label='^2') # graph with dots

plt.title("Our first graph", fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4]) # custom scale for x
# plt.yticks([0,2,4,6,8,10]) # custom scale for y

plt.legend() # shows legend

# save figure
plt.savefig('mygraph.png', dpi=300)

# show plot
plt.show()
```

# Bar chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

## patterns quicker
patterns = ['/','|','\\']
for bar in bars:
    bar.set_hatch(patterns.pop())

## patterns
# bars[0].set_hatch('/')
# bars[1].set_hatch('O')
# bars[2].set_hatch('*')

plt.figure(figsize=(6,4))
plt.show()
```
